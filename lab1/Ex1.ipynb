```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#1
Data = pd.read_csv('Data1.csv')
```

```python
Data
```

- The table above has 1095 rows and 6 columns of random numbers

```python
#2
Data=Data.set_index(Data.columns[0])
Data
```

- First column set as index, so data can be found by the date

```python
#3
cols=['theta_1','theta_2','theta_3','theta_4','theta_5','theta_6']
Data[cols].plot(subplots=True,figsize=(15,10))
plt.show()
```

- Data is presented on the 6 grids with the same date

```python
# 4
Data2=Data.melt()
Data2
```

- Melting the data reshapes it to distinghised variables in columns and assigns the value listed for each of them. It removes dataIndex.

```python
Graph=sns.FacetGrid(Data2, col="variable", sharex=False, col_wrap=3)
Graph.map(plt.hist,"value")
```

- Melting allows to map the values for each variable and create histograms. If one rescales them, they can present probability of occurance.

```python
#5
for i in cols:
    plt.figure()
    sns.distplot(Data[i], kde=True, hist=False)
```

```python
#6
Data2018 = Data['2018-01-01':'2019-01-01']

cols2018 = Data.columns
cols2018 = cols2018[0:4]
print(cols2018)
Reduced2018 = Data2018[cols2018]

Melted2018 = Reduced2018.melt()

Reduced2018.plot(subplots=True,figsize=(15,10))
plt.show()
```

- Subsetting data from one year

```python
plt.figure()
grid = sns.FacetGrid(Melted2018, col="variable")
grid.map(plt.hist, "value")
```
