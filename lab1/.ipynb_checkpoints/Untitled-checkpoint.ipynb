```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Import Data1.csv file to python.
df = pd.read_csv('Data1.csv',index_col=0)
df.head()
```

```python
#Set first column as the index.
#Plot all columns as time series.
for c in df.columns[1:]:
    df[c].plot()
    plt.title(c)
    plt.xticks(rotation=45)
    plt.show()
```

```python
for c in df.columns[1:]:
    df[c].plot()
plt.legend(df.columns[1:])
plt.xticks(rotation=45)
plt.show()
```

```python
#Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot (use sns.FacetGrid and melt, use sharex=False for FacetGrid)
dfm = df.melt()
g = sns.FacetGrid(dfm, col='variable', col_wrap=3, sharex = False)
g = g.map(plt.hist, 'value')
```

```python
#Plot KDE-s (Kernel Denisty Estimators) for all columns.
f = sns.FacetGrid(dfm, col='variable', col_wrap=3, sharex = False)
f = f.map(sns.kdeplot, 'value')
```

```python
#Repeat analysis for columns $\theta_1$-$\theta_4$ in 2018.
df2 = df.loc['2018-01-01':'2018-12-31','theta_1':'theta_4']
df2.head()
```

```python
for c in df2.columns[1:]:
    df2[c].plot()
    plt.title(c)
    plt.xticks(rotation=45)
    plt.show()
```

```python
for c in df2.columns[1:]:
    df2[c].plot()
plt.legend(df.columns[1:])
plt.xticks(rotation=45)
plt.show()
```

```python
df2m = df2.melt()

g = sns.FacetGrid(df2m, col='variable', col_wrap=2, sharex = False)
g = g.map(plt.hist, 'value')

f = sns.FacetGrid(df2m, col='variable', col_wrap=2, sharex = False)
f = f.map(sns.kdeplot, 'value')
```
